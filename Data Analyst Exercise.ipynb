```python
Checkout_information = (pd.merge(checkouts, Libraries, left_on='library_id', right_on='id'))

#Merging the primary key (library_id) from checkouts with the foreign key (id) from Libraries
```

```python
customer_checkout = (pd.merge(Checkout_information, Customers, left_on='patron_id', right_on='id'))

#Merging the primary key (patron_id) from Checkout_information with the foreign key (id) from Customers
```

```python
Customer_checkout = (pd.merge(customer_checkout, books, left_on='id_x', right_on='id'))

#Merging the primary key (id_x) from customer_checkout with the foreign key (id) from books
```

```python
id_x_Customer_checkouts = Customer_checkout.drop(["publishedDate","street_address_x","id_x","authors", "categories", "price", "pages","id_y","street_address_y"], axis = 1)

#Dropping the unformatted columns after replacing them with a formatted version
```

```python
id_x_Customer_checkouts.rename(columns={'patron_id':'Patron_id','title':'Book_title','birth_date':'Birth Date','library_id':'Library_id','name_x':'Library_name','name_y':'Customer_name'})
                               
#Rename and re-format the column headers
```

```python
id_x_Customer_checkouts['Date_checkout_'] = pd.to_datetime(id_x_Customer_checkouts['Date_checkout'],format='%Y%m%d', infer_datetime_format=True)

#Formatting the datetime to a singular readable date format
```

```python
id_x_Customer_checkouts['Date_returned_'] = pd.to_datetime(id_x_Customer_checkouts['Date_returned'],format='%Y%m%d', infer_datetime_format=True)

#Formatting the datetime to a singular readable date format
```

```python
id_x_Customer_checkouts['Days'] = id_x_Customer_checkouts['Date_returned_'] - id_x_Customer_checkouts['Date_checkout_'] 

#Creating the formula to calculate the difference between checkout and returning of library books
```

```python
id_x_Customer_checkouts['Tenor'] = round(pd.to_numeric ((id_x_Customer_checkouts['Date_returned_'] - id_x_Customer_checkouts['Date_checkout_']) / np.timedelta64(1, 'D')) )

#Converting the str "Days" into a numeric value
```

```python
id_x_Customer_checkouts.rename(columns={'patron_id':'Patron_id','title':'Book_title','birth_date':'Birth Date','library_id':'Library_id','name_x':'Library_name','name_y':'Customer_name'})
                               
#Rename and re-format the column headers
```

```python
id_x_Customer_checkouts['Late_Return_Condition'] = np.where(id_x_Customer_checkouts['Tenor'].between(0,28),False,True)

#Set condition to analyze if book was returned within the 28 days period
```

```python
Late_Return_Rate = id_x_Customer_checkouts.groupby(["Late_Return_Condition"])[['Late_Return_Condition']].count()
Late_Return_Rate

#Grouping based on if the books were returned in the permitted time frame
```

```python
Late_Return_Rate = id_x_Customer_checkouts.groupby(["Late_Return_Condition"])[['Late_Return_Condition']].count()
Late_Return_Rate_a = Late_Return_Rate.sum(axis = 0)
```

```python
Late_Return_Rate['Rate'] = Late_Return_Rate/Late_Return_Rate_a

#Calculate the Rate of return as (Total amount late returns / total checkouts)
```
